```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
DATA_PATH = "../data/processed/audio_features.csv"

df = pd.read_csv(DATA_PATH)

df.head()
```

```python
file_names = df["file_name"]

X = df.drop(columns=["file_name"])

print("Feature Shape:", X.shape)
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
X_scaled_df = pd.DataFrame(
    X_scaled,
    columns=X.columns
)

X_scaled_df["file_name"] = file_names

X_scaled_df.head()
```

```python
X_scaled_df = pd.DataFrame(
    X_scaled,
    columns=X.columns
)

X_scaled_df["file_name"] = file_names

X_scaled_df.head()
```

```python
OUTPUT_PATH = "../data/processed/normalized_features.csv"

X_scaled_df.to_csv(OUTPUT_PATH, index=False)

print("Normalized features saved successfully!")
```

## Feature Normalization

Feature normalization is an important preprocessing step in machine learning.

The extracted MFCC, Delta, and Delta-Delta features may have different numerical ranges. If clustering algorithms are applied directly to such data, features with larger values can dominate the distance calculations.

To avoid this issue, we apply **StandardScaler**, which standardizes features using:

z = (x − μ) / σ

where:

μ = mean of the feature  
σ = standard deviation of the feature

After normalization:

• Each feature has mean ≈ 0  
• Each feature has standard deviation ≈ 1  

This ensures that all features contribute equally during clustering.
